<h1> Scatterplot</h1>

```python
fig2, ax2 = plt.subplots(figsize = (20,20))
ax2.scatter(x = TotalJobs,
            y = TotalWomen,
            label = 'Jobs',
            color = 'blue',
            alpha = .5,
            marker = 'd')
ax2.scatter(x = TotalEnrollment, 
            y = EnrollWomen, 
            label = 'Ohio State',
            color = 'red',
            alpha = .5,
            marker = 's')
ax2.set_ylabel('Total Women')
ax2.set_xlabel('Total of people')
ax2.set_title('The relationship between Ohio States enrollment rate\n of women in STEM and the number of women in the STEM field')
ax2.legend()
plt.show()
```
